```python
import xarray as xr

# Path to the raw, unprocessed file you want to check
file_to_check = r'F:\Major Project\Project Working\Datasets\Himalaya_ERA5_Data\2010\SL_2010_08_01_accum.nc'

def check_for_labels(path):
    with xr.open_dataset(path) as ds:
        # tp is usually in meters in raw ERA5, so multiply by 1000 for mm
        max_rain = float(ds.tp.max() * 1000) 
        
        print(f"--- Analysis for {path.split('/')[-1]} ---")
        print(f"Maximum Rainfall found: {max_rain:.2f} mm")
        
        if max_rain >= 100:
            print("RESULT: This file contains Tier 2 (Extreme Rain) candidates.")
        elif max_rain >= 50:
            print("RESULT: This file contains Tier 1 (Heavy Rain) candidates.")
        else:
            print("RESULT: This file ONLY contains Tier 0 (Normal).")

check_for_labels(file_to_check)
```

```python
import xarray as xr
import numpy as np
import os

# Target directory containing your cropped files
target_dir = r'F:\Major Project\Project Working\Datasets\Himalaya_ERA5_Data'

def verify_dataset_quality(directory):
    issues_found = []
    
    # Expected coordinate ranges for Western Himalayas
    EXPECTED_LAT = (28.5, 37.5)
    EXPECTED_LON = (72.0, 81.5)

    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.nc'):
                path = os.path.join(root, file)
                try:
                    # Open without loading into memory for speed
                    with xr.open_dataset(path) as ds:
                        report = {"file": file, "status": "OK", "errors": []}
                        
                        # 1. Check for File Corruption
                        if not ds.data_vars:
                            report["errors"].append("Empty file: No data variables found.")

                        # 2. Verify Coordinate Bounds (Western Himalayas)
                        lat_min, lat_max = ds.latitude.min().item(), ds.latitude.max().item()
                        lon_min, lon_max = ds.longitude.min().item(), ds.longitude.max().item()
                        
                        if not (EXPECTED_LAT[0] <= lat_min <= EXPECTED_LAT[1]) or \
                           not (EXPECTED_LAT[0] <= lat_max <= EXPECTED_LAT[1]):
                            report["errors"].append(f"Latitude out of bounds: {lat_min}-{lat_max}")

                        if not (EXPECTED_LON[0] <= lon_min <= EXPECTED_LON[1]) or \
                           not (EXPECTED_LON[0] <= lon_max <= EXPECTED_LON[1]):
                            report["errors"].append(f"Longitude out of bounds: {lon_min}-{lon_max}")

                        # 3. Check for Missing Data (NaNs)
                        for var in ds.data_vars:
                            null_count = ds[var].isnull().sum().item()
                            if null_count > 0:
                                pct = (null_count / ds[var].size) * 100
                                report["errors"].append(f"Variable '{var}' has {null_count} NaNs ({pct:.2f}%)")

                        # 4. Check for Impossible Values (Outliers)
                        # Example: Temperature (t2m) should be roughly between 220K and 330K
                        if 't2m' in ds.data_vars:
                            t_min, t_max = ds.t2m.min().item(), ds.t2m.max().item()
                            if t_min < 200 or t_max > 350:
                                report["errors"].append(f"Temperature Outlier: {t_min}K - {t_max}K")

                        if report["errors"]:
                            report["status"] = "ISSUE"
                            issues_found.append(report)
                        
                        print(f"Verified: {file} -> {report['status']}")

                except Exception as e:
                    print(f"CRITICAL ERROR opening {file}: {e}")
                    issues_found.append({"file": file, "status": "CORRUPTED", "errors": [str(e)]})

    # Final Summary Report
    print("\n" + "="*50)
    print("FINAL DATA QUALITY REPORT")
    print("="*50)
    if not issues_found:
        print("✅ ALL FILES ARE PERFECT. Correct bounds, no missing data, no outliers.")
    else:
        for issue in issues_found:
            print(f"\n❌ {issue['file']} [{issue['status']}]:")
            for err in issue['errors']:
                print(f"   - {err}")

# Run the verification
verify_dataset_quality(target_dir)
```

```python
import xarray as xr
import os

# Target directory containing your cropped Western Himalayan files
target_dir = r'F:\Major Project\Project Working\Datasets\Himalaya_ERA5_Data'

def drop_metadata_artifacts(directory):
    total_cleaned = 0
    
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.nc'):
                file_path = os.path.join(root, file)
                try:
                    # Open the dataset
                    ds = xr.open_dataset(file_path)
                    
                    # Identify if 'expver' or 'number' exist
                    to_drop = [coord for coord in ['expver', 'number'] if coord in ds.coords or coord in ds.data_vars]
                    
                    if to_drop:
                        # Drop the coordinates
                        # If expver has multiple values, we usually take the first (1) or combine
                        # but for hardening, we simply drop the dimension to flatten the data
                        ds_cleaned = ds.drop_vars(to_drop)
                        
                        # Save back to the file
                        temp_path = file_path + ".clean"
                        ds_cleaned.to_netcdf(temp_path)
                        ds.close()
                        
                        os.replace(temp_path, file_path)
                        print(f"Cleaned {file}: Dropped {to_drop}")
                        total_cleaned += 1
                    else:
                        ds.close()
                
                except Exception as e:
                    print(f"Error cleaning {file}: {e}")

    print(f"\nMetadata cleanup complete. {total_cleaned} files standardized.")

# Run the cleanup
drop_metadata_artifacts(target_dir)
```

```python
import xarray as xr

def open_clean_ds(file_path):
    ds = xr.open_dataset(file_path)
    
    # Check if 'valid_time' exists and rename it to 'time' for consistency
    if 'valid_time' in ds.coords and 'time' not in ds.coords:
        ds = ds.rename({'valid_time': 'time'})
        print(f"Standardized time dimension for {file_path}")
        
    return ds

# Use this function to load any file
ds = open_clean_ds(r'F:\Major Project\Project Working\Datasets\Himalaya_ERA5_Data\2011\SL_2011_06_01_accum.nc')
```

```python
def check_alignment(sl_path, pl_path):
    sl = xr.open_dataset(sl_path)
    pl = xr.open_dataset(pl_path)
    
    # Check 1: Do they have the same time steps?
    if len(sl.valid_time) != len(pl.valid_time):
        print(f"❌ TIME MISMATCH: SL has {len(sl.valid_time)}, PL has {len(pl.valid_time)}")
    
    # Check 2: Do the Lat/Lon grids match exactly?
    if not (sl.latitude.equals(pl.latitude) and sl.longitude.equals(pl.longitude)):
        print("❌ SPATIAL MISMATCH: The grid coordinates do not align between SL and PL.")
    
    # Check 3: Is the 'time' name consistent?
    if 'valid_time' not in sl.coords or 'valid_time' not in pl.coords:
        print("⚠️ NAMING ISSUE: One file is missing the 'valid_time' coordinate name.")
    
    else:
        print("✅ ALIGNMENT PERFECT: Files are ready to be merged for Index Calculation.")

# Test it on your sample
check_alignment(r'F:\Major Project\Project Working\Datasets\Himalaya_ERA5_Data\2011\SL_2011_07_21_inst.nc', r'F:\Major Project\Project Working\Datasets\Himalaya_ERA5_Data\2011\PL_2011_07_21.nc')
```

```python
# Quick check for all files
# If (t2m - d2m) is ever negative, we have a problem.
diff = ds.t2m - ds.d2m
if (diff < -0.01).any():
    print("❌ Physical Error: Dewpoint exceeds Temperature!")
```

```python
# Check if heights are strictly increasing as pressure decreases
# Using our sorted pressure_level (1000 -> 200)
z_diff = ds.z_pl.diff('pressure_level')
if (z_diff < 0).any():
    print("❌ Hydrostatic Error: Geopotential heights are not monotonic!")
```

```python
# Check Surface Temperature (t2m) vs Surface Dewpoint (d2m)
physical_error = ds.t2m < ds.d2m
if physical_error.any():
    print(f"⚠️ Warning: Found {physical_error.sum().values} points where Dewpoint > Temp!")
else:
    print("✅ Surface Thermal Consistency: Passed.")
```

```python
# Check if height increases as we go to higher altitudes (lower pressure)
# Based on your sort (1000 -> 200), diff should be positive
height_diff = ds.z_pl.diff('pressure_level')
if (height_diff < 0).any():
    print("⚠️ Warning: Non-monotonic vertical heights detected!")
else:
    print("✅ Vertical Hydrostatic Consistency: Passed.")
```

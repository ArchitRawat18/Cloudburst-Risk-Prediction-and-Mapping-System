```python
import xarray as xr
import pandas as pd
import os

# --- 1. Load All 4 Components (Targeting July 2011 Event) ---
year, chunk_date = "2011", "2011_07_21"
base = rf'F:\Major Project\Project Working\Datasets\Himalaya_ERA5_Data\{year}'

ds_inst  = xr.open_dataset(os.path.join(base, f'SL_{chunk_date}_inst.nc'))
ds_accum = xr.open_dataset(os.path.join(base, f'SL_{chunk_date}_accum.nc'))
ds_pl    = xr.open_dataset(os.path.join(base, f'PL_{chunk_date}.nc'))
ds_cape  = xr.open_dataset(os.path.join(base, f'SL_{chunk_date}_cape.nc'))

# --- 2. Fix NetCDF Structure Conflicts ---
# Rename 'z' to 'z_pl' to avoid conflict with Surface Geopotential
if 'z' in ds_pl.data_vars:
    ds_pl = ds_pl.rename({'z': 'z_pl'})

# CRITICAL FIX: Ensure pressure_level is indexed for future index-based math
if 'pressure_level' in ds_pl.coords:
    ds_pl = ds_pl.set_index(pressure_level="pressure_level")

# --- 3. Time Alignment & Unit Scaling ---
# Align accumulated data with the instantaneous observation time
ds_accum['valid_time'] = ds_accum.valid_time - pd.Timedelta(minutes=30)

if 'tp' in ds_accum:
    ds_accum['tp'] = ds_accum['tp'] * 1000 # Convert m to mm

# --- 4. Synchronize All Layers to the 30-min Target ---
target_times = ds_accum.valid_time
ds_inst_sync = ds_inst.interp(valid_time=target_times, method='linear')
ds_pl_sync   = ds_pl.interp(valid_time=target_times, method='linear')
ds_cape_sync = ds_cape.interp(valid_time=target_times, method='linear')

# --- 5. Final Master Merge ---
# join='exact' ensures we don't accidentally create mismatched time steps
ds_test = xr.merge([ds_inst_sync, ds_accum, ds_pl_sync, ds_cape_sync], 
                   compat='override', 
                   join='exact')

print("--- HARDENED Synchronized Dataset Created (July 2011) ---")
print(f"Time Range: {ds_test.valid_time.min().values} to {ds_test.valid_time.max().values}")
print(f"Data Variables: {list(ds_test.data_vars)}")
```

```python
import numpy as np
import xarray as xr

# --- 1. PREPARE THE DATASET ---
# FIX: Ensure pressure_level is a dimension index so .sel() works
if 'pressure_level' in ds_test.coords:
    ds_test = ds_test.set_index(pressure_level="pressure_level")

# FIX: Sort pressure levels to ensure vertical integration is positive (925 -> 200)
ds_test = ds_test.sortby('pressure_level', ascending=False)

# Constants 
g = 9.80665 # Gravity

# 2. Helper Function: Calculate Dewpoint (Td)
def calculate_td(t_k, rh):
    t_c = t_k - 273.15
    a, b = 17.625, 243.04
    alpha = np.log(rh/100) + (a * t_c) / (b + t_c)
    td_c = (b * alpha) / (a - alpha)
    return td_c + 273.15

# 3. Extract specific levels
# Now .sel() will work because we set the index above
t850 = ds_test['t'].sel(pressure_level=850.0)
r850 = ds_test['r'].sel(pressure_level=850.0)
t700 = ds_test['t'].sel(pressure_level=700.0)
r700 = ds_test['r'].sel(pressure_level=700.0)
t500 = ds_test['t'].sel(pressure_level=500.0)

td850 = calculate_td(t850, r850)
td700 = calculate_td(t700, r700)

# 4. Thermodynamic Stability Indices
# TT Index
ds_test['TT_index'] = (t850 - t500) + (td850 - t500)

# KI Index: Corrected to Celsius scale
ds_test['KI_index'] = ((t850 - t500) + td850 - (t700 - td700)) - 273.15

# 5. Moisture Parameters
# PWC: FIXED with abs() and sortby to ensure positive water depth
ds_test['pwc'] = abs((ds_test['q'].integrate('pressure_level') * 100) / g)

# VIMT
ds_test['vimt_u'] = (ds_test['q'] * ds_test['u']).integrate('pressure_level') * 100 / g
ds_test['vimt_v'] = (ds_test['q'] * ds_test['v']).integrate('pressure_level') * 100 / g

# 6. Wind and Shear
ds_test['ws_10m'] = np.sqrt(ds_test['u10']**2 + ds_test['v10']**2)
ds_test['shear_u'] = ds_test['u'].sel(pressure_level=500.0) - ds_test['u10']
ds_test['shear_v'] = ds_test['v'].sel(pressure_level=500.0) - ds_test['v10']

print("--- Parameters Successfully Added (Hardened) ---")
print(f"KI Range: {float(ds_test.KI_index.min()):.2f} to {float(ds_test.KI_index.max()):.2f}")
print(f"PWC Range: {float(ds_test.pwc.min()):.2f} to {float(ds_test.pwc.max()):.2f}")
```

```python
# --- 1. Lifted Index (LI) ---
# Hardened: Using the Kelvin variables from our previous block
def approximate_parcel_t500(t850_k, td850_k):
    # Standard linear approximation for tropical/monsoon moisture profiles
    return (t850_k - ((t850_k - td850_k) * 0.2)) - 25  

t500_parcel = approximate_parcel_t500(t850, td850) # Variables from previous block
ds_test['LI_index'] = t500 - t500_parcel

# --- 2. SWEAT Index (Hardened for Units) ---
# Conversion factor: m/s to knots
ms_to_knots = 1.94384

# Wind Speeds in Knots
ws850_kts = np.sqrt(ds_test['u'].sel(pressure_level=850)**2 + 
                    ds_test['v'].sel(pressure_level=850)**2) * ms_to_knots
ws500_kts = np.sqrt(ds_test['u'].sel(pressure_level=500)**2 + 
                    ds_test['v'].sel(pressure_level=500)**2) * ms_to_knots

# Term 1: 850hPa Dewpoint (Set to 0 if Td < 0Â°C)
term1 = 12 * np.maximum(td850 - 273.15, 0)

# Term 2: Total Totals (Set to 0 if TT < 49)
term2 = 20 * np.maximum(ds_test['TT_index'] - 49, 0)

# Term 3 & 4: Wind terms (2*ws850 + ws500)
term3 = 2 * ws850_kts
term4 = ws500_kts

# Note: We omit the 'Shear Term' (Term 5) as it requires complex direction checks
# which are often unreliable in high-terrain ERA5 data.
ds_test['SWEAT_index'] = term1 + term2 + term3 + term4

print("--- HARDENED Complex Parameters Added ---")
print(f"LI Min/Max: {float(ds_test.LI_index.min()):.2f} / {float(ds_test.LI_index.max()):.2f}")
print(f"SWEAT Min/Max: {float(ds_test.SWEAT_index.min()):.2f} / {float(ds_test.SWEAT_index.max()):.2f}")
```

```python
import numpy as np

print("--- HARDENED HEALTH CHECK: JULY 2011 SAMPLE ---")

# 1. Coordinate & Dimension Verification
print(f"Dimensions: {dict(ds_test.dims)}")
expected_vars = ['TT_index', 'KI_index', 'pwc', 'LI_index', 'SWEAT_index', 'tp']

# 2. Physics & Unit Validation
print(f"{'Parameter':<12} | {'Min':<8} | {'Max':<8} | {'NaNs':<6} | {'Status'}")
print("-" * 55)

for p in expected_vars:
    if p in ds_test.data_vars:
        v_min = float(ds_test[p].min())
        v_max = float(ds_test[p].max())
        # We expect a few NaNs at the very start due to the 30-min shift
        nan_count = int(np.isnan(ds_test[p]).sum())
        
        # Logic check based on Himalayan meteorology
        status = "OK"
        if p == 'pwc' and v_min < 0: status = "ERR: Negative"
        if p == 'KI_index' and v_max > 100: status = "ERR: Units (K?)"
        if nan_count > 15000: status = "ERR: High Data Loss" # >1 full time step
        
        print(f"{p:<12} | {v_min:8.2f} | {v_max:8.2f} | {nan_count:<6} | {status}")
    else:
        print(f"{p:<12} | MISSING")

# 3. Rohtang Event Pulse Check (Targeting 32.3N, 77.1E)
# Let's see if our indices peaked during the event buildup
rohtang_buildup = ds_test.sel(
    latitude=32.3, longitude=77.1, 
    valid_time="2011-07-21T19:00:00", # ~12:30 AM IST July 22
    method='nearest'
)
print("\n--- ROHTANG EVENT (JULY 21, 19:00 UTC) PULSE ---")
print(f"SWEAT Index: {float(rohtang_buildup.SWEAT_index):.2f} (Expected > 200)")
print(f"PWC:         {float(rohtang_buildup.pwc):.2f} mm")
print(f"KI Index:    {float(rohtang_buildup.KI_index):.2f}")
```

```python
import xarray as xr
import numpy as np
import pandas as pd
import os

# 1. Load the file
file_path = r'F:\Major Project\Project Working\Datasets\Himalaya_ERA5_Data\2011\Processed_2011_07_21.nc'
ds = xr.open_dataset(file_path)

# Extract filename outside the f-string to avoid SyntaxError
file_name = os.path.basename(file_path)

def run_comprehensive_check(ds, name):
    print(f"=== FINAL HEALTH REPORT: {name} ===\n")
    
    # --- A. METADATA & DIMENSIONS ---
    print("1. STRUCTURE CHECK")
    dims = dict(ds.sizes)
    print(f"   - Dimensions: {dims}")
    
    # We want to ensure 'expver' is gone; 'number' is often redundant
    forbidden = ['expver', 'number'] 
    for f in forbidden:
        if f in ds.coords or f in ds.data_vars:
            print(f"   - [!] WARNING: '{f}' still present in dataset. (Safe but untidy)")
    
    # --- B. NAN DENSITY ANALYSIS ---
    print("\n2. DATA COMPLETENESS (NaN Check)")
    # Total spatial grid size for one time step
    grid_size = dims['latitude'] * dims['longitude']
    
    for var in ds.data_vars:
        nan_count = int(ds[var].isnull().sum())
        nan_pct = (nan_count / (dims['valid_time'] * grid_size)) * 100
        
        # Expect ~0.4% NaNs (exactly 1 time step) due to the 30-min sync shift
        status = "PASS" if nan_pct < 1.0 else "FAIL: High Data Loss"
        print(f"   - {var:<12}: {nan_count:>6} NaNs ({nan_pct:>4.2f}%) -> {status}")

    # --- C. PHYSICAL RANGE VALIDATION ---
    print("\n3. PHYSICAL RANGE VALIDATION")
    # Define realistic bounds for Himalayan Monsoon conditions
    bounds = {
        't2m': (260, 315),       # Kelvin
        'tp': (0, 200),          # mm/hr
        'KI_index': (-100, 60),  # Celsius scale
        'pwc': (0, 100),         # mm
        'SWEAT_index': (0, 800), # Standard units
        'risk_label': (0, 3)     # 0:Normal, 1:H, 2:E, 3:Cloudburst
    }
    
    for var, (vmin, vmax) in bounds.items():
        if var in ds.data_vars:
            actual_min = float(ds[var].min())
            actual_max = float(ds[var].max())
            if actual_min < vmin or actual_max > vmax:
                print(f"   - [!] {var:<11}: Range ({actual_min:.1f} to {actual_max:.1f}) | OUT OF BOUNDS")
            else:
                print(f"   - {var:<12}: Range ({actual_min:.1f} to {actual_max:.1f}) | PASS")

    # --- D. TEMPORAL CONTINUITY ---
    print("\n4. TEMPORAL CONTINUITY")
    time_diffs = np.diff(ds.valid_time.values) / np.timedelta64(1, 'h')
    if all(time_diffs == 1.0):
        print("   - Time Steps: Consistent 1-hour intervals | PASS")
    else:
        print(f"   - [!] Time Steps: Gaps detected! Intervals: {np.unique(time_diffs)}")

    # --- E. LABEL INTEGRITY ---
    print("\n5. LABEL DISTRIBUTION")
    labels, counts = np.unique(ds.risk_label.values, return_counts=True)
    for lbl, count in zip(labels, counts):
        desc = {0:"Normal", 1:"Heavy", 2:"Extreme", 3:"Cloudburst"}.get(lbl, "Unknown")
        print(f"   - Tier {lbl} ({desc:<10}): {count:>6} points")

run_comprehensive_check(ds, file_name)
ds.close()
```
